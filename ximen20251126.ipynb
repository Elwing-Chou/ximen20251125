````python
import random
import math

# x^2/4 + y^2/9 = 1
a, b = 2, 3
# 準備一個把它包起來的正方形
w, h = 10, 10

count = 0
times = 1000000

i = 0
while i < times:
    halfw, halfh = w / 2, h / 2
    x = random.uniform(-1*halfw, halfw)
    y = random.uniform(-1*halfh, halfh)
    z = x ** 2 / a ** 2 + y ** 2 / b ** 2
    if z <= 1:
        count = count + 1
    i = i + 1

ratio = count / times
area = ratio * (w * h)
print("猜的是:", area)

real_area = math.pi * a * b
print("我用公式算:", real_area)
````

````python
import random
# 1. 判斷時間很自由
# 2. 收尾很自由
# 3. 條件撰寫自由
trans = ["even", "win", "lose"]
win, lose = 0, 0
while True:
    result = random.randint(0, 2)
    print(trans[result])
    if result == 1:
        win = win + 1
    elif result == 2:
        lose = lose + 1

    print(win, ":", lose)
    if win >= 3 or lose >= 3:
        if abs(win - lose) < 2:
            print("DEUCE")
        else:
            break
````

````python
def is_leap_year(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

def is_date_valid(date_checked):
    day = date_checked % 100
    remain = date_checked // 100
    month = remain % 100
    year = remain // 100

    if year > 0:
        if 12 >= month >= 1:
            day_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            max_day = day_per_month[month-1]
            if month == 2 and is_leap_year(year):
                max_day = max_day + 1

            if 1 <= day <= max_day:
                return True
    return False

max_count = 100
start = 4499
count = 0
while True:
    # 把數字把平方當成要檢查的日期
    d = start * start
    if is_date_valid(d):
        print(d)
        count = count + 1
        if count == max_count:
            break
    start = start + 1
````
