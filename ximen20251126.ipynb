````python
total = 0
# xx階乘
factorial = 1

i = 0
while i < 10000:
    # 第一次不能*0
    if not i == 0:
        factorial = factorial * i
    # 1/0! + 1/1!
    total = total + 1 / factorial
    i = i + 1

print(total)
````

````python
import random
import math

# x^2/4 + y^2/9 = 1
a, b = 2, 3
# 準備一個把它包起來的正方形
w, h = 10, 10

count = 0
times = 1000000

i = 0
while i < times:
    halfw, halfh = w / 2, h / 2
    x = random.uniform(-1*halfw, halfw)
    y = random.uniform(-1*halfh, halfh)
    z = x ** 2 / a ** 2 + y ** 2 / b ** 2
    if z <= 1:
        count = count + 1
    i = i + 1

ratio = count / times
area = ratio * (w * h)
print("猜的是:", area)

real_area = math.pi * a * b
print("我用公式算:", real_area)
````

````python
import random
# 0: 平手 1: 營 2: 書
trans = ["even", "win", "lose"]
win, lose = 0, 0
# cond1: win < 3 cond2: lose < 3
while (win < 3) and (lose < 3):
    result = random.randint(0, 2)
    print(trans[result])
    if result == 1:
        win = win + 1
    elif result == 2:
        lose = lose + 1

if win > lose:
    print("win!!!!!!")
else:
    print("lose!!!!!!!")
````

````python
import random
# 1. 判斷時間很自由
# 2. 收尾很自由
# 3. 條件撰寫自由
trans = ["even", "win", "lose"]
win, lose = 0, 0
while True:
    result = random.randint(0, 2)
    print(trans[result])
    if result == 1:
        win = win + 1
    elif result == 2:
        lose = lose + 1

    if win == 3:
        print("win!!!!!!")
        break
    elif lose == 3:
        print("lose!!!!!!")
        break
````
