````python
i = 0
while i < 10:
    n1 = i + 1
    n2 = 10 - i
    n3 = 2 * i
    print("第", n1, n2, n3, "次")
    i = i + 1
````

````python
total = 0

i = 0
while i < 10:
    total = total + (i + 1)
    print(total)
    i = i + 1
````

````python
total = 0
# xx階乘
factorial = 1

i = 0
while i < 10000:
    # 第一次不能*0
    if not i == 0:
        factorial = factorial * i
    # 1/0! + 1/1!
    total = total + 1 / factorial
    i = i + 1

print(total)
````
