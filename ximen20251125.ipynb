````python
# calculate bmi: 體重(kg) / 身高(m) ^ 2
weight = float(input("請輸入體重(kg):"))
height = float(input("請輸入身高(cm):"))
bmi = weight / (height / 100) ** 2

print("你輸入的體重是:" + str(weight))
print("你輸入的身高是:" + str(height))
print("計算出的bmi是:" + str(bmi))

# 縮排 : TAB
# 過濾單選題: 嚴格放上面
if bmi >= 25:
    print("over")
elif bmi >= 18:
    print("normal")
else:
    print("under")

print("!!!!")
````

````python
# import .py名字
import random

# 清單: ["x", "x", "x"]
# python會幫你做一個對照編號(key): 0 1 2
# key-value查詢操作: 查東西[key]
trans = ["剪刀", "石頭", "布"]

my = int(input("0.剪刀 1.石頭 2.布:"))
# 照抄import後面的東西  .:的
com = random.randint(0, 2)
print("我出:", trans[my])
print("電腦出:", trans[com])

if my == (com + 1) % 3:
    print("我贏")
elif com == (my + 1) % 3:
    print("電腦贏")
else:
    print("平手")
````
